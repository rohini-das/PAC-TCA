```python
def all_subsets(AA):
    
    def powerset(A):
        length = len(A)
        return {
        frozenset({e for e, b in zip(A, f'{i:{length}b}') if b == '1'})
        for i in range(2 ** length)}

    s = set()
    s = powerset(AA)
    l = list(s)
    l1 = [set(x) for x in l]
    #print(list(sub_sets))
    return l1
```
